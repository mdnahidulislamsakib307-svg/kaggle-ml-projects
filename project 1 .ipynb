```python
corr = df['Hours_Studied'].corr(df['Exam_Score'])
print(corr)
plt.scatter(corr,corr.index,color='blue',figsize=(15,
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('scatter plot :Hours Studis by Exam score ')
plt.show()
```

```python
grouped = df.groupby('Parental_Involvement')['Exam_Score'].mean()
print(grouped)
grouped.plot(kind='bar',color=['blue','green','red'],edgecolor='black',figsize=(15,5))
plt.title('Average exam score by parental involvement')
plt.ylabel('Average exam score')
plt.show()
```
